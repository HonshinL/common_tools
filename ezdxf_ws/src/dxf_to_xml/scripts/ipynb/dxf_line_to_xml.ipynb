```python
import ezdxf
import xml.etree.ElementTree as ET

def dxf_line_to_xml(dxf_file, xml_file):
    doc = ezdxf.readfile(dxf_file)
    msp = doc.modelspace()

    shape = ET.Element("shape")
    waypoints = ET.SubElement(shape, "waypoints")

    index = 1
    for line in msp.query("LINE"):
        start = line.dxf.start
        end = line.dxf.end

        ET.SubElement(waypoints, "waypoint",
                      index=str(index), x=str(start.x), y=str(start.y), z=str(start.z or 0),
                      r="0", speed="5")
        index += 1

        ET.SubElement(waypoints, "waypoint",
                      index=str(index), x=str(end.x), y=str(end.y), z=str(end.z or 0),
                      r="0", speed="8")
        index += 1

    tree = ET.ElementTree(shape)
    ET.indent(tree, space="    ")   # 添加缩进和换行
    tree.write(xml_file, encoding="utf-8", xml_declaration=True)
    print(f"已导出XML文件: {xml_file}")

# 示例调用
dxf_line_to_xml("line.dxf", "line.xml")
```
